```python
#manual selecting of school type and dist
options1=driver.find_element_by_xpath("//select[@name='SchoolType']/option[text()='ASSOCIATED SCHOOL']")
options1.click()
options2=driver.find_element_by_xpath("//select[@name='Dist1']/option[text()='AMRITSAR']").click()
driver.find_element_by_css_selector("input[type='submit']").click()
psoup=bs4.BeautifulSoup(driver.page_source)
table=psoup.find('table',id='t01')
print(table.get_text())
```

```python
#hovering 
elements=driver.find_elements_by_class_name("tip")
for element in elements:
  hover=ActionChains(driver).move_to_element(element)
  hover.perform()
  text=element.get_attribute('rel')
  #text=WebDriverWait(driver,10).until(EC.presence_of_element_located((By.XPATH,"//[@class='tip']")))
  print(text)
  time.sleep(1)
```

```python
#full code (manual)
rawtext=''
hovertext=''
options1=driver.find_element_by_xpath("//select[@name='SchoolType']/option[text()='ASSOCIATED SCHOOL']")
options1.click()
options2=driver.find_element_by_xpath("//select[@name='Dist1']/option[text()='FATEHGARH SAHIB']").click()
driver.find_element_by_css_selector("input[type='submit']").click()
psoup=bs4.BeautifulSoup(driver.page_source)
table=psoup.find('table',id='t01')
if table!=None:
    rows=driver.find_elements_by_xpath("//table/tbody/tr")
    while len(rows)>2:
         rawtext=rawtext+table.get_text()
         elements=driver.find_elements_by_class_name("tip")
         for element in elements:
             #hover=ActionChains(driver).move_to_element(element)
             #hover.perform()
             hovertext=hovertext+element.get_attribute('rel')
             #time.sleep(1)
         driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
         b=driver.find_element_by_id("btnNext")
         b.click()
         rows=driver.find_elements_by_xpath("//table/tbody/tr")
print('DONE')
```

```python
#automatic search of all school and dist
radio =driver.find_element_by_id("SchoolType")
options1=radio.find_elements_by_tag_name("option")
radio1 =driver.find_element_by_id("Dist1")
options2=radio1.find_elements_by_tag_name("option")
for option in options1[1:]:
    option.click()
    for i in range(1,len(options2)):
            print(i)
            driver.execute_script("window.scrollTo(0,0)")
            options2[i].click()
            driver.find_element_by_css_selector("input[type='submit']").click()
```
