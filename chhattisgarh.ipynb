```python
import pandas as pd
df=pd.read_excel('chhattisgarh.xlsx')
df.head()
```

```python
import pandas as pd
schoolname=[]
coordinator = []
email=[]
phno1=[]
for i in df['Name of the Institution/ School']:
    schoolname.append(i)
for i in df['Coordinator (Center\rIncharge for SCERT CG)']:
    coordinator.append(i)
for i in df['Email ID']:   
    email.append(i)
for i in df['Mobile No.']:   
    phno1.append(i)
print(len(schoolname))
#print(coordinator)
```

```python
import csv
outputFile = open('partial.csv', 'w', newline='')
outputWriter = csv.writer(outputFile)

outputWriter.writerow(['school name', 'principal name','contact no','email'])
#outputFile.close()
for i in range (len(schoolname)):
    scname=schoolname[i]
    princiname=coordinator[i]
    email1=email[i]
    phno=phno1[i]
    outputWriter.writerow([scname,princiname,phno,email1])
```
