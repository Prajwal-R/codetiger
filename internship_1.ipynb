```python
import requests,bs4,webbrowser,csv

new_file='file.csv'
res=requests.get('https://www.cisce.org/')
#res.raise_for_status()
bs=bs4.BeautifulSoup(res.text,features="html.parser")

```

```python
locate=bs.select('.top-nav a')
r=requests.get('https://cisce.org/'+locate[0].get('href'))
#r.raise_for_status()
b=bs4.BeautifulSoup(r.text,features="html.parser")

show=b.select('input[value="Show All"]')
re=requests.get('https://cisce.org/locate-search.aspx?country=0&state=0&dist=0&city=0&location=&schooltype=&cve=&isc=&icse=&schoolclassi=&school=&search=locate')
re.raise_for_status()
soup=bs4.BeautifulSoup(re.text,features="html.parser")

```

```python
table=soup.find('table')
for i,tr in enumerate(table.findAll('tr')):
    row=[]
    for td in tr.findAll('td'):
        row.append(td.text)
        if i==0:
            with open(new_file,'w') as f:
                writer=csv.DictWriter(f,row)
                writer.writeheader()
        else:
            with open(new_file,'a') as f:
                 writer=csv.writer(f)
                 writer.writerow(row)
```

```python
name=[]
for i in range (len(words)-1):
    word=words[i+1].split(',')
    name.append(word[0])
print(name)
```

```python
sclname=[]
for i in range (len(name)):
    school=name[i].split(' ')
    sname=''
    for j in range(len(school)-1):
        #print(school[j+1])
        sname+=school[j+1]+' '
    sname=sname.strip()
    sclname.append(sname)
print(sclname)
```
