```python
sclname=[]
for i in range (len(name)):
    school=name[i].split(' ')
    sname=''
    for j in range(len(school)-1):
        #print(school[j+1])
        sname+=school[j+1]+' '
    sname=sname.strip()
    sclname.append(sname)
print(sclname)
```
